```python
import pandas as pd

TOKEN = "01a2fc8b129f46c4978784cca0c54421"
url = "https://api.football-data.org/v4/competitions/CL/standings"
r = requests.get(url, headers={"X-Auth-Token": TOKEN})
data = r.json()

# Example: print the league-phase table
for table in data.get("standings", []):
    if table.get("type") in ("TOTAL", "LEAGUE_PHASE"):  # depends on season format
        for row in table["table"]:
            print(f'{row["position"]:>2}. {row["team"]["name"]} '
                  f'P:{row["playedGames"]} W:{row["won"]} D:{row["draw"]} L:{row["lost"]} '
                  f'GF:{row["goalsFor"]} GA:{row["goalsAgainst"]} GD:{row["goalDifference"]} '
                  f'Pts:{row["points"]}')

club_names = [
    row["team"]["name"]
    for table in data.get("standings", [])
    if table.get("type") in ("TOTAL", "LEAGUE_PHASE")
    for row in table["table"]
]

# --- 1) Actual table: make a dict club -> actual rank (1-based) ---

actual_rank = {name: i + 1 for i, name in enumerate(club_names)}
# Example: {'FC Bayern München': 1, 'Arsenal FC': 2, ...}

# --- 2) Read your friends' predictions (already standardized names) ---

file_path = "Final_Predictions_standardized.CSV"
Predictions = pd.read_csv(file_path)

# --- 3) Compute score for each friend ---

scores = {}  # friend name -> total points

for friend in Predictions.columns:
    total_diff = 0
    
    # go through predicted table from 1st to 36th
    for predicted_pos, club in enumerate(Predictions[friend], start=1):
        actual_pos = actual_rank.get(club)
        if actual_pos is None:
            # in case of typo / missing club, just warn and skip
            print(f"Warning: {club!r} in column {friend!r} not found in actual standings")
            continue
        
        diff = abs(predicted_pos - actual_pos)
        total_diff += diff
    
    scores[friend] = total_diff

# --- 4) Show ranking of friends (fewest points = winner) ---

scores_series = pd.Series(scores, name="score").sort_values()
print(scores_series)

winner = scores_series.idxmin()
print("\nWinner:", winner, "with", scores_series.min(), "points")
```

```python
import pandas as pd

TOKEN = "01a2fc8b129f46c4978784cca0c54421"
url = "https://api.football-data.org/v4/competitions/CL/standings"
r = requests.get(url, headers={"X-Auth-Token": TOKEN})
data = r.json() 

club_names = [
    row["team"]["name"]
    for table in data.get("standings", [])
    if table.get("type") in ("TOTAL", "LEAGUE_PHASE")
    for row in table["table"]
]

# 1) actual standings from API (club_names already built)
actual_rank = {name: i + 1 for i, name in enumerate(club_names)}
# e.g. {'FC Bayern München': 1, 'Arsenal FC': 2, ...}

# 2) read predictions
file_path = "Final_Predictions_standardized.CSV"
Predictions = pd.read_csv(file_path)

# 3) scoring with bonuses
scores = {}  # friend -> total score

for friend in Predictions.columns:
    print("---------------------------")
    print(friend)
    total_points = 0
    
    for predicted_pos, club in enumerate(Predictions[friend], start=1):
        actual_pos = actual_rank.get(club)
        if actual_pos is None:
            print(f"Warning: {club!r} in column {friend!r} not found in actual standings")
            continue
        
        # base difference
        diff = abs(predicted_pos - actual_pos)

        # bonus (negative = good)
        bonus = 0
        if predicted_pos <= 8 and actual_pos <= 8:
            bonus -= 3          # correct top-8 team
        elif predicted_pos <= 24 and actual_pos <= 24:
            bonus -= 1          # correct 9–24 team
        print(club)
        print(diff + bonus)
        total_points += diff + bonus
    
    scores[friend] = total_points

# 4) ranking of friends (lowest score wins)
scores_series = pd.Series(scores, name="score").sort_values()

print("---------------------------")
print("---------------------------")

print(scores_series)

winner = scores_series.idxmin()
print("\nWinner:", winner, "with", scores_series.min(), "points")
```
